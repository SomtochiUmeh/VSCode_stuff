Given a list `names` of donuts available and a list `weights` of their respective weights, help Pelle find out what donuts he should consider buying. 

**Task:** Write a function called `donuts(names, weights)` that accepts 2 lists as described above and returns a list, containing the names of donuts that Pelle should consider, along with the weights of those donuts (see format below).


<b>Input format:</b> 
- `names`: a *list* containing the names (*str*) of the donuts available at the shop.
- `weights`: a *list* containing the weights (*int* or *float*) of each donut in oz.

<b>Output format:</b> *list* with structure `[name1, weight1, name2, weight2, ...]` containing the information of the donuts that weigh at least 5 oz.

````python
# TODO: You can put your code here:
def donuts(names, weights):
    # create a list to store values
    lst = []   

    # a for loop to check each item in names and their respective weights in weights
    for name, weight in zip(names, weights):
        # if the weight chacked exceeds 5
        if weight>=5:
            # add both name and weight to lst
            lst.extend((name, weight))
    # return final list
    return lst
````

````python
# You can run this cell to import the data
import pandas as pd

# Location of the .csv file
url = "http://tinyurl.com/postsummer2021-task6-csv2"

# Loading the file into DataFrame
df = pd.read_csv(url)

# Display top 5 rows from the table
df.head()
````

````python
# TODO: You can put your code here:
def scores_to_mode(year1, year2):
    
    # function to find mode of scores in a year
    def find_mode(year):
        # creating a dataframe of the scores
        df_1 = df[df['Year'] == year]
        # creating a list of the grade column
        scores = list(df_1['Grade'])
        
        # initializing a variable to store the mode
        mode = 0
        # initializing a count variable to store the max number of times a grade appears in the scores list
        count = 0

        # a for loop to iterate through the range of possible scores [1,2,3,4]
        for num in range(1, 5):
            # if the number of times a grade appears exceeds the current count
            if scores.count(num) > count:
                # set the count variable
                count = scores.count(num)
                # set the mode
                mode = num
        # returns mode after iteration is complete
        return mode
        
    # returns a boolean stating if year1's mode is greater than year2's
    return find_mode(year1) > find_mode(year2)
````

In this exercise, you will be using the random and matplotlib libraries in Python to write a **simulation** of the Marbles Game. The rules of the game are as follows: 
* There is a bag with 3 marble balls: 1 blue, 1 green, and 1 red. A blue marble is equal to 1 point, a green marble is 3 points, and a red marble is 5 points.
* There are at least 2 players.
* On each turn, a player draws two marbles from the bag, with replacement. 
* If a single green marble is drawn, the player scores nothing, and the turn ends.
* If two blue marbles are drawn, the player‚Äôs entire score is reset, and the turn ends.
* Otherwise, the sum of the marbles is added to the player‚Äôs total score, and the turn ends.
* Whoever accumulates 23 points in the fewest amount of turns wins.
* If more than 1 player accumulates 23 points or more in the same amount of turns, then the game ends in a tie, and all players with 23 points or more win.

Write a function called `marbles_game(players)` that inputs the number of players (the maximum number of players is 4) and simulates the game. **You can accept that the players are not playing simultaneously, but are playing one after the other.**  This means that Player 1 plays until they reach 23 points, then Player 2 plays until they reach 23 points, etc.

````python
# We prepared for you some usefull libraries. Just RUN this cell
import matplotlib.pyplot as plt
import random

%matplotlib inline
````

````python
# TODO: You can put your code here:
def marbles_game(players):
    
    # checks if the number of players is within range
    if players not in range(2, 5):
        return 'Minumum number of players is 2, maximum is 4'
        
    # variable that stores the winning number of turns set to a very high initial number
    winning_turns = 500000
    # varible storing the winner
    winner = 0

    # for loop for each of the players to play
    for player in range(1, players+1):
        
        # variable storing the total number of turns
        turns = 0
        # list storing the scores
        score = [0]
        
        # an infinite loop than will be broken is a condition is met within the loop
        while True:
            
            # 3 for green, 1 for blue, 5 for red
            player_turn = random.choices([3, 1, 5], k=2)
            turns += 1

            # if a single green marble is pickes
            if 3 in player_turn and player_turn[0] != player_turn[1]:
                # no score, append previous value on scores list
                score.append(score[-1])
            # if two blue balls are picke
            elif player_turn[0] == player_turn[1] == 1:
                # reset score, append 0 to scores list
                score.append(0)
            else:
                # add values of the marble to the score
                score.append(sum(player_turn)+ score[-1])

            # breaks the loop when a player reaches winning point
            if score[-1] >= 23:
                break

        # if the number of turns is less than winning_turns, we have a current winner
        if len(score)<winning_turns:
            # update winner
            winner = player
            # update winning_turns
            winning_turns = len(score)

        # if we have more than one winner
        elif len(score)==winning_turns:
            # update winner
            winner = f'{winner} and {player}'
            # update winning_turns
            winning_turns = len(score)

        # plot points against turns
        plt.plot(range(turns+1), score)

    # decorate the plot
    plt.legend(['1', '2', '3', '4'], loc='lower right')
    plt.xlabel('turns')
    plt.ylabel('points')
    plt.title('Marbles Game')

    plt.show()
    # return winner
    return f'Player {winner} won'
````

````python
# TODO: You can put your code here:
def simulate_news(n):
    # first day
    day = 1
    # one person has heard
    heard = 1
    # everyone else hasn't heard
    not_heard = n-1
    
    # list to store number of people that hear the news over time
    lheard = [heard]
    # list storing number of people who haven't heard over time
    lnot_heard = [not_heard]
    
    # loop that continue while there is at least one person who hasn't heard the news
    while lnot_heard[-1]>0:
        # calculating the new people that hear using the probability given and giving space for realistic devistions in the data
        new_heard = 0.03*not_heard + random.uniform(-2,2)
        # update heard and not_heard
        heard += new_heard
        not_heard -= new_heard
        
        # append updates values to their respective lists and increment day by one
        lheard.append(heard)
        lnot_heard.append(not_heard)
        day += 1
        
    # plot how the news spread against the number of days
    plt.plot(range(1, day+1), lheard)
    plt.plot(range(1, day+1), lnot_heard)

    # decorate the plot
    plt.xlabel('Days')
    plt.ylabel('Number of People')
    plt.title('Minervaplois News Spread')
    plt.legend(['Know the news', 'Do not know the news'], loc='center right')
    plt.show()
    
    # return number of days it took
    return day
````
